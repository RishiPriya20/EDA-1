**Handle missing values**

```python
missing_values = df.isnull().sum()
df.fillna(df.mean(), inplace=True)
```

**Detect and handle outliers**

```python
import numpy as np
df = df[(np.abs(df - df.mean()) <= (3*df.std())).all(axis=1)]
```

**Correlation heatmap**

```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

**Observation**

This code generates a heatmap to visualize the correlation matrix of the dataset's numerical variables
The 'df.corr()' function calculates the pairwise correlation coefficient between the numerical columns in the data frame.
The 'sns.heatmap()' function from 'seaborn' library is then used to create a heatmap of this correlation matrix 'annot = True ' displays the correlation values within the heatmap and cmap=collwarm provides a color gradient for better visualization and 'plt.show()' is used to display the heatmap()
